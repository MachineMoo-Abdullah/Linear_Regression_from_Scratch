```python
def loss_function(m,b,points):
  total_error = 0
  for i in range(len(points)):
    x = points.iloc[i].StudyTime
    y = points.iloc[i].Score
    total_error += (y - (m*x + b))

  total_error = total_error / float(len(points))

def gradient_descent(m_now,b_now,points,learning_rate):
  m_gradient = 0
  b_gradient = 0

  n = len(points)
  for i in range(len(points)):
    x = points.iloc[i].StudyTime
    y = points.iloc[i].Score

    m_gradient += -(2/n) * x * (y - (m_now * x + b_now))
    b_gradient += -(2/n) * (y - (m_now * x + b_now))

  m_new = m_now - m_gradient * learning_rate
  b_new = b_now - b_gradient * learning_rate

  return m_new, b_new

m = 0
b = 0
learning_rate = 0.0001
epochs = 300

for i in range(epochs):
  if i % 50 == 0:
    print("Epochs : ",i)
  m,b = gradient_descent(m,b,df,learning_rate)

plt.scatter(df.StudyTime,df.Score)
plt.plot(list(range(0,100)),[m * x + b for x in list(range(0, 100))],color = 'red')
plt.show()

print(f"({m},{b})")
```
